# План исследования



**Заказчик**: IT-компания, последние 15 лет успешно занимавшаяся разработкой и консалтингом на российском рынке.

**Методология исследования**: Методология CRISP-DM.

**Этапы исследования**:

1.	Оценка полученного датасета на предмет соответствия цели анализа.
2.	Сбор данных существующих онлайн-курсов.
3.	Анализ продуктовой линейки и стоимости программ дополнительного образования.
4.	Сопоставительный анализ данных программ дополнительного образования конкурентов.
5.	Анализ предлагаемых курсов на предмет соответствия реальным требованиям работодателей, предъявляемым к соискателям.  
6.	Анализ баз вакансий.
7.	Оценка результатов исследования.
8.	Подготовка отчетов.

**Бизнес-цель заказчика**: Создать собственный онлайн-курс, который будет актуальным для целевой аудитории, а также востребованным на рынке труда. Выйти на рынок онлайн-курсов.

**Объект исследования**: 
* Рынок онлайн-обучения
* Рынок труда

**Предмет исследования**: 
1.  Существующие курсы онлайн-обучения: 
* тематическая направленность курсов; 
* стоимость курсов; 
* продолжительность обучения; 
* содержание курсов;
* количество программ; 
* оценка и отзывы.
2.  Вакансии IT-специалистов: 
* наиболее востребованные навыки;
* наиболее востребованные вакансии;
* заработная плата.

**Цель анализа данных**:
* выявить актуальные направления подготовки IT-специалистов;
* изучить конкурентов;
* сформировать рекомендации по ценообразованию курсов.

**Требования к результату анализа**:
* Наглядное описание ситуации на рынке труда и ситуации в дополнительном профессиональном образовании (онлайн-курсы);
* Результат бенчмаркинга продуктов конкурентов;
* Представление выводов и рекомендаций: какие образовательные продукты нужны рынку, какую цену (или диапазон цен) можно назначить, чтобы не продешевить, но и люди могли покупать;
* Визуализация полученных результатов (дашборд);
* Отчёт по результатам исследования;
* Оформление результатов исследования в виде презентации.

**Планируемые источники данных**:

* продукция конкурентов: агрегатор онлайн-курсов obrazoval.ru;
* вакансии рынка труда: база вакансий России - hh.ru.

**Типы данных**:
* полуструктурированные данные;
* открытые данные.

**Способ(-ы) получения данных**:
* данные вакансий получены от заказчика, собранные с помощью агрегатора;
* данные онлайн-курсов собраны с помощью агрегатора.

**Методы анализа**:
1.  Описательная статистика:
* количество вакансий; 
* средняя заработная плата;  
* анализ аудитории;
* описание ситуации на рынке труда и онлайн-курсов; 
* выводы и рекомендации. 
2.  Сравнительный анализ: 
* бенчмаркинг продуктов конкурентов.
3.  Классификация:
* анализ продуктовой линейки.

**Инструменты анализа**:
Google-таблицы, MS Excel, Google Colab, язык программирования Python, Power BI

**Риски и условия реализации проекта**:
* недостаточный объем данных; 
* плохое качество данных;
* данные не подходят под задачи;
* отсутствие закономерностей в данных;
* падение покупательской способности.
